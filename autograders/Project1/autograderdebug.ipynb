```python
# Define function to compute whether or not values are within a certain acceptable error
def is_within_tolerance(studVal, solnVal, tolerance):
    #studVal is a number, solnVal is a number, tolerance is a decimal between 0 and 1
    if abs(studVal-solnVal) <= tolerance*solnVal:
        return True
    else:
        return False
```

```python


def grade_decision_matrix(file_path):
    # Load student-submitted Excel file
    df = pd.read_excel(file_path, header=1, index_col=0)
    
    # Extract raw computed values
    computed_values = df.iloc[:, ::3]  # Every third column contains computed values
    computed_values.columns = ['Efficiency', 'Cost', 'Volume']
    
    # Normalize each column
    normalized_values = computed_values.apply(normalize, axis=0)
    
    # Define weights
    weights = {'Efficiency': 1, 'Cost': 4, 'Volume': 2}
    
    # Compute weighted scores
    weighted_values = normalized_values.copy()
    for col in weighted_values.columns:
        weighted_values[col] = compute_weighted(normalized_values[col], weights[col])
    
    # Compare student-submitted values for VoltEdge and Ampere Dynamics
    correct = True
    for vendor in ['VoltEdge Solutions', 'Ampere Dynamics']:
        for metric in ['Efficiency', 'Cost', 'Volume']:
            student_norm = df.loc[vendor, f'{metric} Normalized']
            student_weighted = df.loc[vendor, f'{metric} Weighted']
            correct_norm = normalized_values.loc[vendor, metric]
            correct_weighted = weighted_values.loc[vendor, metric]
            
            if not (abs(student_norm - correct_norm) < 0.01 and abs(student_weighted - correct_weighted) < 0.01):
                correct = False
    
    # Compute expected values for "Your Design"
    correct_norm_your_design = normalized_values.loc['Your Design']
    correct_weighted_your_design = weighted_values.loc['Your Design']
    
    # Calculate total score
    score = 1.0 if correct else 0.0
    
    # Gradescope output
    results = {
        "score": score,
        "output": "Autograder completed.",
        "visibility": "visible",
        "execution_time": "N/A",
        "tests": []
    }
    
    # Save results to JSON
    with open("results.json", "w") as f:
        json.dump(results, f)
    
if __name__ == "__main__":
    grade_decision_matrix(sys.argv[1])
```
